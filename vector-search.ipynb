```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd
```

```python
books = pd.read_csv("books_cleaned.csv")
books
```

```python

books["tagged_description"]
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
# Use HuggingFace for free embeddings
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
# Store in Chroma vector DB
db_books = Chroma.from_documents(documents, embedding=embedding)
```

```python
query = "A book about time travel and space exploration"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
isbn_str = docs[0].page_content.split()[0].strip().replace('"', '')
books[books["isbn13"] == int(isbn_str)]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```
